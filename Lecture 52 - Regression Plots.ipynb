```python
import pandas as pd
import numpy as np
from numpy.random import randn
import warnings
warnings.filterwarnings('ignore')
from scipy import stats
import matplotlib as mpt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
# Specify what order fit you want

sns.lmplot('total_bill','tip',tips,order=4,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips
```

```python
tips['tip_pect']=(tips['tip']/tips['total_bill'])*100
tips
```

```python
sns.lmplot('size','tip_pect',tips)
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
sns.lmplot('size','tip_pect',tips,hue='sex',markers=['o','x'])
```

```python
sns.lmplot('size','tip_pect',tips,hue='day')
```

```python
sns.regplot('total_bill','tip_pect',tips)
```

```python
fig, (axis1,axis2) = plt.subplots(2,1,sharey=True)

sns.regplot('total_bill','tip_pect',tips,ax=axis1)
sns.violinplot(x='size',y='tip_pect',data=tips,color='#C6011F',ax=axis2)
```
