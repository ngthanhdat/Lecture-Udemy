```python
import pandas as pd
import numpy as np
from numpy.random import randn
import warnings
warnings.filterwarnings('ignore')
from scipy import stats
import matplotlib as mpt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1,data2])
```

```python
# Handle outliers

sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
sns.boxplot(x=data1)
```

```python
#Normal Distribution

data_1 = stats.norm(0,5).rvs(100)

# Two gamma Dist. concatenated together

data_2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])

# Box plot both data_1 and data_2

sns.boxplot(data=[data_1,data_2],whis=np.inf)
```

```python
sns.violinplot(data=[data_1,data_2])
```

```python
sns.violinplot(data=data_2,bw=0.01)
```

```python
sns.violinplot(data=data_1,inner='stick')
```
